```python
#fasttext可以用于文本分类
#文本分类的过程：获取数据 训练集和验证集划分 训练 评估 调参 保存
```

```python
import fasttext
#训练模型
model = fasttext.train_supervised(input = './data/cooking.train')
model.save_model('./data/1.bin')
```

```python
# 加载模型
model = fasttext.load_model('./data/1.bin')

# 预测
result = model.predict('Why not put knives in the dishwasher ?')
print(result)
```

```python
#查看在验证集上的表现
result2 = model.test('./data/cooking.valid')
print(result2)
```

```python
#数据清洗后训练效果有所提升
model1 = fasttext.train_supervised(input = './data/cooking_pre.train')
model1.save_model('./data/2.bin')
```

```python
# 加载模型
model = fasttext.load_model('./data/2.bin')

# 预测
result3 = model.predict('Why not put knives in the dishwasher ?')
print(result3)
result4 = model.test('./data/cooking_pre.valid')
print(result4)
```

```python
#模型调优 
#2.增加训练轮数  调整学习率  增加ngram特征  修改损失计算方式
#自动超参数调优
model2 = fasttext.train_supervised(input = './data/cooking_pre.train',autotuneValidationFile = './data/cooking_pre.valid',autotuneDuration = 60)
model2.save_model('./data/3.bin')
```

```python
#实际生活中loss = ova修改损失方式
model3 = fasttext.train_supervised(input = './data/cooking_pre.train',lr = 0.5,epoch = 25,wordNgrams = 2,loss = 'ova')
model3.save_model('./data/3.bin')
```

```python
model = fasttext.load_model('./data/3.bin')
result5 = model.predict('Why not put knives in the dishwasher ?',k=-1, threshold = 0.5 )
print(result5)
result6 = model.test('./data/cooking_pre.valid')
print(result6)
```
