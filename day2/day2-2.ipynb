```python
import torch
import jieba
import torch.nn as nn
from IPython import embed
from tensorflow.keras.preprocessing.text import Tokenizer
from tinycss2 import tokenizer
from torch.utils.tensorboard import SummaryWriter
import tensorflow as tf
import tensorboard as tb
# 修正文件系统接口
tf.io.gfile = tb.compat.tensorflow_stub.io.gfile

def dm02_nnembeding_show():
    sentence1 = '传智教育是一家上市公司，旗下有黑马程序员品牌。我是在黑马这里学习人工智能'
    sentence2 = "我爱自然语言处理"
    sentences = [sentence1, sentence2]  # 列表
    word_list = []
    for i in sentences:
        word_list.append(jieba.lcut(i))
        # print(word_list)
    # 获得word_index index_word
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(word_list)
    print(tokenizer.word_index)
    print(tokenizer.index_word)

    # 将文本转化为一个数字
    seq = tokenizer.texts_to_sequences(word_list)
    print(seq)

    # 获取样本中所有单词
    words = tokenizer.word_index.keys()
    print(words)

    # 创建nn.Embedding层 实例化
    embed = nn.Embedding(num_embeddings=len(words), embedding_dim=8)  # 前者为需要进行词向量表示的单词总个数  后者代表每个单词进行词嵌入的维度
    print(embed.weight.shape)

    # 创建SummarWriter
    #summarywriter = SummaryWriter()
    #summarywriter.add_embedding(embed.weight.data, words)  # 前者为每个词代表的向量
    #summarywriter.close()

     #可视化

    for idx in range(len(tokenizer.index_word)):
        tmpvec = embed(torch.tensor(idx))
        print('%4s' %(tokenizer.index_word[idx + 1]), tmpvec.detach().numpy())

if __name__ == '__main__':
    dm02_nnembeding_show()
```
