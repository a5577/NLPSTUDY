```python
#获取形容词的函数
def get_a_list(text):
    r = []
    for g in pseg.lcut(text):#g为词加词性
        if g.flag == 'a':#词性
            r.append(g.word)#获取词
    return r
    
def get_word_cloud(keywords_list):
     # 实例化词云生成器对象
    wordcloud = WordCloud(max_words=100, background_color='white')
    # 准备数据
    keywords_string = " ".join (keywords_list)#列表变成一个字符串
    # 产生词云
    wordcloud.generate(keywords_string)
    # 画图
    plt.figure()
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis('off')
    plt.show()
    
def dm_word_cloud():
    train_data = pd.read_csv(filepath_or_buffer = './train.tsv', sep='\t')#按照制表符分割
    p_train_data = train_data[train_data['label']==1]['sentence']#获取正样本
    #获取正样本中的形容词
    p_a_train_vocab = list(chain(*map(lambda x: get_a_list(x),p_train_data)))
    get_word_cloud(p_a_train_vocab)
dm_word_cloud()
```
