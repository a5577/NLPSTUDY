```python
#获取标签数量分布
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 读训练集 验证集数据pd.read_csv 
train_data = pd.read_csv(filepath_or_buffer = './train.tsv', sep='\t')#按照制表符分割
dev_data = pd.read_csv(filepath_or_buffer = './dev.tsv', sep='\t')
sns.countplot(x='label',data=train_data, hue = 'label')
plt.title('train_label')
plt.show()
```

```python
# 验证集上标签的数量分布
# 3-2 sns.countplot() 统计label标签的0、1分组数量
sns.countplot(x='label', data = dev_data,hue='label')

# 4-2 画图展示 plt.title() plt.show()
plt.title('dev_label')
plt.show()
```
