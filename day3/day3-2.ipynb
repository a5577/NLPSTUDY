```python
#-------------------------长度统计分析------------------------------
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
 # 1 设置显示风格plt.style.use('fivethirtyeight')
plt.style.use('fivethirtyeight')

# 2 pd.read_csv 读训练集 验证集数据
train_data = pd.read_csv(filepath_or_buffer='./train.tsv', sep='\t')
dev_data = pd.read_csv(filepath_or_buffer='./dev.tsv', sep='\t')

# 3 求数据长度列 然后求数据长度的分布map将可迭代对象中的每个元素都经过函数处理得到新的map对象（迭代器）
train_data['sentence_length'] = list(map(lambda x: len(x), train_data['sentence']))
dev_data['sentence_length'] = list(map(lambda x: len(x), dev_data['sentence ']))

    # 4 统计正负样本长度散点图 （对train_data数据，按照label进行分组，统计正样本散点图）
sns.countplot(x='sentence_length', data=train_data,hue='sentence_length')
plt.show()
sns.countplot(x='sentence_length', data=dev_data,hue='sentence_length')
plt.show()
```

```python
#散点图
sns.stripplot(y='sentence_length', x='label', data=train_data)
plt.show()

sns.stripplot(y='sentence_length', x='label', data=dev_data)
plt.show()
```
