```python
#--------------------获取不同词汇总数统计-------------------
import jieba
import pandas as pd
from itertools import chain
```

```python
train_data = pd.read_csv(filepath_or_buffer = './train.tsv', sep='\t')#按照制表符分割
dev_data = pd.read_csv(filepath_or_buffer = './dev.tsv', sep='\t')
train_vocab = set(chain(*map(lambda x: jieba.lcut(x),train_data['sentence'])))
print(len(train_vocab))
dev_vocab = set(chain(*map(lambda x: jieba.lcut(x),dev_data['sentence '])))
print(len(dev_vocab))
```
