```python
#-------------------------文本的特征处理-----------------------
#给定一段文本序列, 其中n个词或字的相邻共现特征即n-gram特征, 常用的n-gram特征是bi-gram和tri-gram特征, 分别对应n为2和3.
```

```python
#zip
list1 = [1,2,3,5]
list2 = [2,3,4,5]
list3 = [3,4,5,6,7,8]
a = zip(list1,list2,list3)
b,c,d = zip(*a)#解压缩
print(b)
print(c)
print(d)
```

```python
# 一般n-gram中的n取2或者3, 这里取2为例
ngram_range = 2
def create_ngram_set(input_list):
    return set(zip(*[input_list[i:] for i in range(ngram_range)]))
input_list = [1,2,5,8,9,5]
res = create_ngram_set(input_list)
print(res)
```

```python
#文本的长度规范
#将句子统一长度  不足截断  用0补齐
from tensorflow.keras.preprocessing import sequence
culten = 10#句子长度
def padding(x_train):
    return sequence.pad_sequences(x_train,culten,padding = 'post',truncating= 'post')#补齐默认从前面补齐 从后面补齐用post  后者为截断
x_train = [[1,2,3,5,4,8,5,6,45,5,5,8,5,45,8,8],[5,5,5,8,8,8]]
res = padding(x_train)
print(res)
```
