```python
#分词jieba
import jieba
content = "传智教育是一家上市公司，旗下有黑马程序员品牌。我是在黑马这里学习人工智能"
# 精确模型：试图将句子最精确地切开，适合文本分析。也属于默认模式
jieba.cut(content, cut_all=False)  # cut_all默认为False
 # 将返回一个生成器对象
```

```python
jieba.lcut(content, cut_all=False)#返回列表
```

```python
jieba.lcut(content, cut_all=True)
```

```python
import jieba
def cutnum():
    content = "传智教育是一家上市公司，旗下有黑马程序员品牌。我是在黑马这里学习人工智能"
    #返回生成器 取出元素的方法：for next list
    result = jieba.cut(content,cut_all=False)
    print(result)
    for result1 in result:
        print(result1)
cutnum()
```

```python
import jieba
sentence = '传智教育是一家上市公司，旗下有黑马程序员品牌。我是在黑马这里学习人工智能'
mydata = jieba.lcut(sentence ,cut_all=False)
print(mydata)
jieba.load_userdict("./userdict.txt")#加载自定义词典
mydata2 = jieba.lcut(sentence,cut_all=False)#根据定义的词典里面的词进行分词 但是要考虑词频
print(mydata2)
#词频越大就可以拆分开
```
